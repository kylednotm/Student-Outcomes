```python
import pandas as pd
import time
```

```python
def cds_generation(year_frame):
    local_cds = []
    for i in range (1, len(year_frame)):
        try:
            local_cds.append((str(int(year_frame.loc[i, "CountyCode"]))+str(int(year_frame.loc[i, "DistrictCode"]))+str(int(year_frame.loc[i, "CharterNumber"]))+str(int(year_frame.loc[i, "SchoolCode"]))))
        except:
            local_cds.append(None)
    local_cds.append(None)        
    return local_cds
```

```python
CAHSEE_df = [0]
page=12

start = time.time()
print("Starting Job 1")

for i in range(page, page+1):
    if i<10:
        year="0"+str(i)
    else:
        year=str(i)
        
    print("appending year 20"+year)
    
    if i == 3 or i == 4:
        data = pd.read_csv('Resources/CAHSEE/CAHSEE20'+year+'.txt', sep=',', engine='python')
    else:    
        data = pd.read_csv('Resources/CAHSEE/CAHSEE20'+year+'.txt', sep='\\t', engine='python')
            
    CAHSEE_df.append(data)

for column in CAHSEE_df[1]:
    if column[0] == ' ':
        CAHSEE_df[1].rename(columns={column:column[1::]}, inplace=True)
    
print("Job 1 Successfully Completed!")
end = time.time()
print("Total time: " + str(end - start) + " seconds")
print('-'*20)

start = time.time()
print("Starting Job 2")

cds = cds_generation(CAHSEE_df[1])

print("Job 2 Successfully Completed!")
end = time.time()
print("Total time: " + str(end - start) + " seconds")
print('-'*20)

CAHSEE_df[1]['CDS']=cds
CAHSEE_df[1].head()
```

```python
SAT_df = [0]

start = time.time()
print("Starting Job 1")
for i in range(1, 2):

    if i<10:
        year="0"+str(i)
    else:
        year=str(i)    
    data = pd.read_excel('Resources/SAT_ACT_AP/sat'+year+'.xls')
    SAT_df.append(data)
    
    SAT_df[i]['YEAR']=int(str(20)+str(year))

print("Job 1 Successfully Completed!")
print("Starting Job 2")

SAT_columns_full = []
year=0

for page in SAT_df[1:14]:
    SAT_columns_local = []
    year+=1
    if year < 9:
        for column in SAT_df[year].iloc[SAT_df[year].index==1].values.tolist()[0]:
            SAT_columns_local.append(column.replace('\n', ' '))
        page.columns=SAT_columns_local
        SAT_df[year] = SAT_df[year].drop(SAT_df[year].index[0:2])
        SAT_df[year] = SAT_df[year].reset_index(drop=True)
        
    elif year == 9:
        for column in SAT_df[year].iloc[SAT_df[year].index==2].values.tolist()[0]:
            SAT_columns_local.append(column.replace('\n', ' '))
        page.columns=SAT_columns_local
        SAT_df[year] = SAT_df[year].drop(SAT_df[year].index[0:3])
        SAT_df[year] = SAT_df[year].reset_index(drop=True)

    elif year == 10:
        for column in SAT_df[year].iloc[SAT_df[year].index==3].values.tolist()[0]:
            SAT_columns_local.append(column.replace('\n', ' '))
        page.columns=SAT_columns_local
        SAT_df[year] = SAT_df[year].drop(SAT_df[year].index[0:4])
        SAT_df[year] = SAT_df[year].reset_index(drop=True)

    elif year < 14:
        for column in SAT_df[year].iloc[SAT_df[year].index==2].values.tolist()[0]:
            SAT_columns_local.append(column.replace('\n', ' '))
        page.columns=SAT_columns_local
        SAT_df[year] = SAT_df[year].drop(SAT_df[year].index[0:3])
        SAT_df[year] = SAT_df[year].reset_index(drop=True)
        
SAT_column_names = ['County Number', 'District Number', 'School Number', 'County Name',
       'District Name', 'School Name', 'Grade 12 Enroll ment', 'Number Tested',
       'Percent Tested', ' Critical Reading Average', ' Math Average',
       ' Writing Average', ' Total Average', 'Total >=1500 Number',
       'Total >=1500 Percent']

for page in SAT_df[6:13]:
    page.columns=SAT_column_names
    
for page in SAT_df[1:6]:
    page['Writing Average'] = None
    
for page in SAT_df[1:13]:
    page.rename(columns={'Total >=1500 Number':'Total High SAT Score',
                              'Number w/Score >=1000':'Total High SAT Score',
                              'Total >=1500 Percent':'Percent High SAT Score',
                              'Percent w/Score >=1000':'Percent High SAT Score'},
               inplace=True)
    
    for column in page:
        if column[0] == ' ':
            page.rename(columns={column:column[1::]}, inplace=True)
    
    page['CDS_CODE'] = ''     
    for index, row in page.iterrows():
        page.loc[index, 'CDS_CODE']=(str(int(row["County Number"]))+str(int(row["District Number"]))+str(int(row["School Number"])))

print("Job 2 Successfully Completed!")
end = time.time()
print("Total time: " + str(round(end - start, 2)) + " seconds")
SAT_df[1]
```
